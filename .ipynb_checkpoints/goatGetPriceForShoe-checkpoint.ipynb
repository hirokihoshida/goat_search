```python
import requests
import json
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import numpy as np
```

```python
SEARCH_TERM=''


#Get top shoe IDs
url = 'https://2fwotdvm2o-dsn.algolia.net/1/indexes/product_variants_v2/query?x-algolia-agent=Algolia for vanilla JavaScript 3.25.1&x-algolia-application-id=2FWOTDVM2O&x-algolia-api-key=ac96de6fef0e02bb95d433d8d5c7038a'
data = {'distinct':'true', 'hitsPerPage':'50','page':'0', 'query': SEARCH_TERM}
r = requests.post(url, data=json.dumps(data))
shoeList = pd.DataFrame.from_records(r.json()['hits'])
shoeList = shoeList[['product_template_id', 'name', 'main_picture_url']]
shoeList.rename(columns={'product_template_id': 'id', 'main_picture_url': 'picture'}, inplace=True)
shoeList
```

```python
shoeDict = {}
for _, shoe in shoeList.iterrows():
    print('Processing', shoe['name'])
    #Get shoe info
    url = 'https://www.goat.com/web-api/v1/product_variants'
    param = {'productTemplateId':shoe['id']} 
    header = {'User-Agent': 'Mozilla/5.0'}
    req = requests.get(url, headers=header, params=param)
    jsonInfo = req.json()
    #Turn to dataframe and clean up
    s = pd.DataFrame.from_records(jsonInfo)
    s.drop(['sizeOption', 'lowestPriceCentsByWarehouse', 'instantShipLowestPriceCents'], axis=1, inplace=True)
    s.rename(columns={'lowestPriceCents': 'price'}, inplace=True)
    s['price'] = [re.findall('(\d+)00', str(x))[0] for x in s['price']]
    s['price'] = s['price'].astype(int)
    shoeDict[shoe['name']] = s
```

```python
for name, df in shoeDict.items():
    print(name)
    p = df[(df.shoeCondition != 'new_no_defects')]
    display(p)
```

```python
availableSizes = p['size'].unique()
```

```python
#take in table of prices of 1 shoe, return with price differences

#sortType 0 = used vs newNoDefect; 1 = newNoDefect vs newWithDefect; 2 = newNoDefectGoodBox vs newNoDefectBadBox
#boxCond 0 = no preference; 1 = box exists in good condition; 2 = box exists
def sizeSort(shoes, sortType, boxCond, name):
    output = dict()
    
    #used vs newNoDefect
    if sortType == 0:
        category1 = 'new_no_defects'
        category2 = 'used'
    #newNoDefect vs newWithDefect
    elif sortType == 1:
        category1 = 'new_no_defects'
        category2 = 'new_with_defects'
        
    availableSizes = shoes['size'].unique()
    for size in availableSizes:
            sizeSelect = shoes.loc[shoes['size']==size]
            if (category1 not in sizeSelect['shoeCondition'].values) or (category2 not in sizeSelect['shoeCondition'].values):
                continue
            #no preference in box
            if boxCond == 0:
                price1 = sizeSelect.loc[sizeSelect['shoeCondition'] == category1]['price'].sort_values().iloc[0]
                price2 = sizeSelect.loc[sizeSelect['shoeCondition'] == category2]['price'].sort_values().iloc[0]
            #box exists in good condition
            elif boxCond == 1:
                price1 = sizeSelect.loc[sizeSelect['shoeCondition'] == category1]
                if 'good_condition' not in price1['boxCondition'].values:
                    continue
                price2 = sizeSelect.loc[sizeSelect['shoeCondition'] == category2]
                if 'good_condition' not in price2['boxCondition'].values:
                    continue
                price1 = price1.loc[sizeSelect['boxCondition'] == 'good_condition']['price']
                price2 = price2.loc[sizeSelect['boxCondition'] == 'good_condition']['price']
            #box exists
            elif boxCond == 2:
                price1 = sizeSelect.loc[sizeSelect['shoeCondition'] == category1]
                if (('good_condition' not in price1['boxCondition'].values) 
                    and ('badly_damaged' not in price1['boxCondition'].values)
                    and ('missing_lid' not in price1['boxCondition'].values)):
                    continue
                price2 = sizeSelect.loc[sizeSelect['shoeCondition'] == category2]
                if (('good_condition' not in price2['boxCondition'].values)
                    and ('badly_damaged' not in price2['boxCondition'].values)
                    and ('missing_lid' not in price2['boxCondition'].values)):
                    continue
                price1 = price1['price'].sort_values().iloc[0]
                price2 = price2['price'].sort_values().iloc[0]
            output[price1 - price2] = name, str(size)
    return output
            


#sortType 0 = used vs newNoDefect; 1 = newNoDefect vs newWithDefect; 2 = newNoDefectGoodBox vs newNoDefectBadBox
#boxCond 0 = no preference; 1 = box exists in good condition; 2 = box exists
priceChart = sizeSort(shoeDict["Air Jordan 1 Retro High OG 'Crimson Tint'"], sortType=1, boxCond=0, name="Air Jordan 1 Retro High OG 'Crimson Tint'")
bestDifference = np.max(list(priceChart.keys()))
print(bestDifference, priceChart[bestDifference])
```

```python
for name, df in shoeDict.items():
    priceChart = sizeSort(df, sortType=1, boxCond=0, name=name)
    if len(priceChart) == 0:
        continue
    bestDifference = np.max(list(priceChart.keys()))

    print(bestDifference, priceChart[bestDifference])
```
